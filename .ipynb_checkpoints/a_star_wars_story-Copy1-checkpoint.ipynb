```python
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url_people = "http://swapi.dev/api/people/" 
url_planets = "http://swapi.dev/api/planets/" 
url_films = "http://swapi.dev/api/films/" 
url_species = "http://swapi.dev/api/species/" 
url_vehicles = "http://swapi.dev/api/vehicles/" 
url_starships = "http://swapi.dev/api/starships/"
```

```python
response_people = requests.get(url_people).json()
print(json.dumps(response_people, indent=2, sort_keys=True))
```

```python
character_names = []

for name in (response_people['results']):
    names_response = requests.get(name).json()
    names_all = names_response['name']
    character_names.append(names_all)

print(character_names)
```

```python
with open(response_people) as data_file:
    data = json.load(data_file)
    for v in data.values():
        print(v['x'], v['y'], ['z'])
```
